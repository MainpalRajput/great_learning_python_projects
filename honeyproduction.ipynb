## 1. Import required libraries and read the dataset.

```python
try:
    import numpy as np 
    import pandas as pd 
    import seaborn as sns
    import matplotlib.pyplot as plt
    import warnings
    warnings.filterwarnings("ignore")
    DeprecationWarning("ignore")
except FileNotFoundError as err:
    print(err__name__)    
```

```python
#function for read file
def read_file(file_loca,file_name):
    try:
        return pd.read_csv(file_loca+"/"+file_name)
    except:
        raise FileNotFoundError    
     
```

## 3. Display the percentage distribution of the data in each year using the pie chart.

```python
#percentage distribution of the data in each year using pie chart
plt.figure(figsize= (12,9))        
df["year"].value_counts().plot(kind = 'pie',autopct =" %1.2f%%")
plt.title('Percentage Distribution for year')
plt.show()
```

 the code efficiently produces a well-labeled pie chart that visually conveys the percentage distribution of data among various years. The use of Pandas for data manipulation and Matplotlib for visualization makes the process streamlined, resulting in a clear representation of the data's temporal distribution.

## 5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings.

## 6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings

```python
#plot boxplot using seaborn
plt.figure(figsize=(15,8))
sns.boxplot(df,x="year",y="prodvalue")
plt.show()
```

By examining the box plot, you can identify differences in production value distributions across different years. You can observe whether there are any significant variations or trends in production values for each year. This visualization helps you quickly grasp how the production values are distributed and whether there are any potential anomalies or patterns.we can see that outlier present in all years.
